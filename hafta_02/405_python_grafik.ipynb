## Python ile grafik oluşturma

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # grafikleri web tarayıcısında görüntülemek için gerekli olabilir
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
print(x,y)
```

## Matplotlib ile grafik çizimi

```python
plt.plot(x, y)
```

```python
plt.plot(x,y,'r') # 'r' : red (kırmızı renk)
plt.xlabel('X değerleri')
plt.ylabel('Y değerleri')
plt.title('y = x^2 grafiği')
plt.show()
```

```python
y3 = x**3 
```

```python
y3
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, label="x^2")
ax.plot(x, y3, label="x^3")
ax.legend()
```

## Seaborn ile grafik oluşturma

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Dağılım grafiği
sns.distplot(tips['total_bill'])
```

```python
# jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# pairplot
sns.pairplot(tips)
```

```python
# barplot
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# countplot
sns.countplot(x='sex',data=tips)
```

```python
# boxplot
sns.boxplot(x='day', y='total_bill',data=tips, palette='rainbow')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips)
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

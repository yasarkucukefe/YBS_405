## Boyut İndirgeme (Dimensionality Reduction)
#### Bir veri setinin yüksek boyutlu bir uzaydan daha düşük bir boyutlu uzaya indirgenmesi (çok kolondan oluşan bir veri seti üzerinde analiz yapmak amacıyla daha az sayıda kolon sayısına ingirgenmesi). Bu işlemde bir anlam kaybı olur ancak makul düzeyde kaldığı sürece kabul edilebilir. Bir dönüştürme algoritması ile gerçekleştirilir.
https://en.wikipedia.org/wiki/Dimensionality_reduction

## 1. PCA (Principle Component Analysis) (Temel Bileşen Analizi)

```python
# https://scikit-learn.org/stable/

# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA # PCA kütüphanesi
```

```python
trn = pd.read_csv('https://raw.githubusercontent.com/yasarkucukefe/YBS_405/main/data/optdigits_trn.csv')
```

```python
trn.head() # Eğitim veri seti (training)
```

```python
tst = pd.read_csv('https://raw.githubusercontent.com/yasarkucukefe/YBS_405/main/data/optdigits_tst.csv')
```

```python
tst.head() # Test seti (test)
```

```python
# Eğitim setinin ayrılması X-y
X_trn = trn.filter(regex='\d') # Kolon isminde sayısal bir değer olan kolonları filtrele
y_trn = trn.y # y kolonunu al
```

```python
# Test setinin ayrılması X-y
X_tst = tst.filter(regex='\d') # Kolon isminde sayısal bir değer olan kolonları filtrele
y_tst = tst.y # y kolonunu al
```

```python
X_trn.head()
```

```python
X_tst.head()
```

## PCA Model

```python
pca = PCA() # PCA nesnesini oluştur
```

```python
# Transformasyon (Transformation)
trn_tf = pca.fit_transform(X_trn)
```

```python
X_trn.head()
```

```python
print(trn_tf)
```

```python
type(trn_tf)
```

```python
trn_tf[:,1].mean()
```

```python
X_trn['P1'].mean()
```

```python
trn_tf[:,1].std()
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
print(pca.score(X_trn))
```

```python
tst_tf = pca.transform(X_tst)
```

```python
print(tst_tf)
```

```python
sns.scatterplot(x=trn_tf[:,0],y=trn_tf[:,1],style=y_trn,hue=y_trn,palette=['red','green','blue'])
```

```python
sns.scatterplot(x=tst_tf[:,0],y=tst_tf[:,1],style=y_tst,hue=y_tst,palette=['red','green','blue'])
```

```python
print(pca.score(X_tst))
```

## Linear Discriminant Analysis (LDA) (Doğrusal Diskriminant Analizi)
https://en.wikipedia.org/wiki/Linear_discriminant_analysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
trn = pd.read_csv('https://raw.githubusercontent.com/yasarkucukefe/YBS_405/main/data/optdigits_trn.csv')
tst = pd.read_csv('https://raw.githubusercontent.com/yasarkucukefe/YBS_405/main/data/optdigits_tst.csv')
```

```python
# Eğitim setinin ayrılması X-y
X_trn = trn.filter(regex='\d') # Kolon isminde sayısal bir değer olan kolonları filtrele
y_trn = trn.y # y kolonunu al
# Test setinin ayrılması X-y
X_tst = tst.filter(regex='\d') # Kolon isminde sayısal bir değer olan kolonları filtrele
y_tst = tst.y # y kolonunu al
```

```python
# LDA Model
lda = LinearDiscriminantAnalysis()
lda.fit(X_trn,y_trn)
```

```python
trn_tf = lda.transform(X_trn)
```

```python
sns.scatterplot(x=trn_tf[:,0],y=trn_tf[:,1],style=y_trn,hue=y_trn,palette=['red','green','blue'])
```

```python
tst_tf = lda.transform(X_tst)
sns.scatterplot(x=tst_tf[:,0],y=tst_tf[:,1],style=y_tst,hue=y_tst,palette=['red','green','blue'])
```

```python
# Skorlar
lda.score(X_trn,y_trn)
```

```python
lda.score(X_tst,y_tst)
```

## t-SNE: t-distributed stochastic neighbor embedding
https://en.wikipedia.org/wiki/T-distributed_stochastic_neighbor_embedding

```python
from sklearn.manifold import TSNE
```

```python
trn = pd.read_csv('https://raw.githubusercontent.com/yasarkucukefe/YBS_405/main/data/optdigits_trn.csv')
tst = pd.read_csv('https://raw.githubusercontent.com/yasarkucukefe/YBS_405/main/data/optdigits_tst.csv')
```

```python
# Eğitim setinin ayrılması X-y
X_trn = trn.filter(regex='\d') # Kolon isminde sayısal bir değer olan kolonları filtrele
y_trn = trn.y # y kolonunu al
# Test setinin ayrılması X-y
X_tst = tst.filter(regex='\d') # Kolon isminde sayısal bir değer olan kolonları filtrele
y_tst = tst.y # y kolonunu al
```

```python
# perplexity=1
tsne = TSNE(n_components=2, perplexity=1, random_state=1)
```

```python
X_trn_tf = tsne.fit_transform(X_trn)
```

```python
sns.scatterplot(x=X_trn_tf[:,0],y=X_trn_tf[:,1],style=y_trn,hue=y_trn,palette=['red','green','blue'])
```

```python
# FARKLI PARAMETRELER
# perplexity=2
tsne = TSNE(n_components=2, perplexity=2, random_state=1)
X_trn_tf = tsne.fit_transform(X_trn)
sns.scatterplot(x=X_trn_tf[:,0],y=X_trn_tf[:,1],style=y_trn,hue=y_trn,palette=['red','green','blue'])
plt.title("Perplexity=2")
```

```python
# perplexity=50
tsne = TSNE(n_components=2, perplexity=50, random_state=1)
X_trn_tf = tsne.fit_transform(X_trn)
sns.scatterplot(x=X_trn_tf[:,0],y=X_trn_tf[:,1],style=y_trn,hue=y_trn,palette=['red','green','blue'])
plt.title("Perplexity=50")
```

```python
X_tst_tf = tsne.fit_transform(X_tst)
sns.scatterplot(x=X_tst_tf[:,0],y=X_tst_tf[:,1],style=y_tst,hue=y_tst,palette=['red','green','blue'])
plt.title("Perplexity=50 - Test Seti")
```
